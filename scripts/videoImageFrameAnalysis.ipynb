```python
# Importing libraries
import os
import cv2
import glob

# creating directory
def create_dir(dir_name):
    try:
        if not os.path.exists(dir_name):
            os.mkdir(dir_name)
            print(dir_name, 'directory created on path', os.getcwd() )
        else:
            raise OSError
    except OSError:
        print(f'Error MSG: Directory already exists')

# remove old files from the directory
def clear_dir(path):
    i = 0
    try:
        file_lst = glob.glob(os.path.join(path,'*.png'))
        if not len(file_lst) == 0:
            for file in file_lst:
                i=i+1
                os.remove(file)
            print(i, 'files removed from', path)
        else:
            raise OSError
    except OSError:
        print(f'ERROR MSG: Folder is empty --->', path)

# Capturing img frames
def capture_frames(video_path,image_frame_path):
    cap_fps = 2 # capturing frames in every 2 seconds
    for videos in os.listdir(video_path):
        cap = cv2.VideoCapture(video_path + videos)
        fps = int(cap.get(cv2.CAP_PROP_FPS)) # no of frames per second to run
        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) # no of image frames 
        print(videos, ' No of frames per second:',fps,'Total number of frames:', frame_count)

    temp = 0    
    for videos in os.listdir(video_path):
        i = 0
        cap = cv2.VideoCapture(video_path + videos)
        for frame_idx in range (int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
            ret,frame = cap.read()
            if frame_idx%cap_fps == 0:
                i += 1
                cv2.imwrite(f'{image_frame_path}{frame_idx}__{videos}.png',frame)
                temp += 1
        print('From {} no of image_frames captured: '.format(videos),i)
    print('Total number of image frames: ',temp)
    


if __name__ == "__main__":
    create_dir('train_frames')
    clear_dir('./train_frames')
    capture_frames('./Avenue Dataset/training_videos/','./train_frames/')

    create_dir('test_frames')
    clear_dir('./test_frames')
    capture_frames('./Avenue Dataset/testing_videos/','./test_frames/')

```
